```python
from transformers import pipeline
import pandas as pd

ner = pipeline(
    "ner",
    model="dslim/bert-base-NER",
    tokenizer="dslim/bert-base-NER",
    aggregation_strategy="simple"
)

text = "Last week, Sundar Pichai, the CEO of Google, attended an international technology conference in San Francisco, California. During the event, he spoke about the future of Artificial Intelligence and announced a new collaboration between Google, Microsoft, and NVIDIA to develop more efficient AI systems. The conference was attended by several well-known industry leaders, including Satya Nadella from Microsoft, Jensen Huang from NVIDIA, and Sam Altman from OpenAI. Representatives from Stanford University and the Massachusetts Institute of Technology also presented their latest research in machine learning and robotics. After the conference, the delegates visited Silicon Valley and held meetings with startups before traveling to New York to attend another business summit. The event attracted participants from countries such as India, Germany, Japan, Canada, and the United Kingdom, making it one of the largest global technology gatherings of the year."

results = ner(text)

entity_list = []

for entity in results:
    entity_list.append({
        "text": entity["word"],
        "type": entity["entity_group"],
    })

# Convert to DataFrame
df = pd.DataFrame(entity_list)

# Display
print(df)
```
