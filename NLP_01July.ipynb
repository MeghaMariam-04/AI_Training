```python
#stemming

import nltk
from nltk.stem import PorterStemmer

nltk.download('punkt')

stemmer = PorterStemmer()

words = ["running", "runner", "played", "playing", "studies"]

for word in words:
    print(word, "->", stemmer.stem(word))
```

```python
#Lemmatization

import nltk
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')
nltk.download('omw-1.4')

lemmatizer = WordNetLemmatizer()

words = ["running", "better", "cars", "feet"]

for word in words:
    print(word, "->", lemmatizer.lemmatize(word))
```

```python
#Bag of Words

from sklearn.feature_extraction.text import CountVectorizer

documents = [
    "Oil production is increasing",
    "Machine learning predicts oil production",
    "Oil production is important"
]

vectorizer = CountVectorizer()

X = vectorizer.fit_transform(documents)

print(vectorizer.get_feature_names_out())

print(X.toarray())
```

```python
#TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer

documents = [
    "Oil production is increasing",
    "Machine learning predicts oil production",
    "Oil production is important"
]

vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(documents)

print(vectorizer.get_feature_names_out())

print(X.toarray())
```

```python
#Word2Vec
from gensim.models import Word2Vec

sentences = [
    ["oil", "production", "is", "high"],
    ["machine", "learning", "predicts", "production"],
    ["artificial", "intelligence", "helps", "oil", "industry"]
]

model = Word2Vec(
    sentences,
    vector_size=50,
    window=3,
    min_count=1,
    workers=1
)

print(model.wv["oil"])

print("\nWords similar to 'oil':")
print(model.wv.most_similar("oil"))
```
